```python
def countCurrency(amount,dn):

    notes = dn
    print ("Currency Count -> ")
    for i in notes:
        c=0
        s=notes[i]
        while amount <i and s>0:
            c=c+1
            s=s-1
     #   ditres[i]=c
            amount = amount-i
        print (i ," : ", notes[i])


dn={2000:10,  500:10,  100:20, 50:0, 20:10, 10:5,  5:10, 2:10, 1:10}
amount = 1256
countCurrency(amount,dn)
```

```python
def breakdown(amount,dn):

    notes = dn
    print ("Currency Count -> ")
    for i in notes:
        c=0
        s=notes[i]
        while amount >=i and s>0:
            c=c+1
            s=s-1
            amount = amount-i
        print (i ," : ", c)


dn={2000:10,  500:10,  100:20, 50:0, 20:10, 10:5,  5:10, 2:10, 1:10}
amount = 1256
breakdown(amount,dn)
```
